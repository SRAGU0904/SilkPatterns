```python
import pandas as pd
import re
```

```python
data = pd.read_excel('data//SilkPattern_Song&Yuan_original.xlsx')
```

```python
# Remove spaces + remove content inside parentheses (both English () and Chinese （）)
def clean_text(x):
    if isinstance(x, str):
        # Remove leading/trailing spaces and all internal spaces
        x = x.strip().replace(" ", "")
        # Remove content inside English () and Chinese （）
        x = re.sub(r"\(.*?\)|（.*?）", "", x)
    return x

# Apply to the entire DataFrame
data = data.applymap(clean_text)
```
